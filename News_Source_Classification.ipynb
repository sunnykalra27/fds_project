```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import layers, Model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.utils.class_weight import compute_class_weight

print("TensorFlow version:", tf.__version__)

# 1) Load dataset
df = pd.read_csv("NewFolder.csv")
print("Original shape:", df.shape)
print(df.head())

# 2 Basic text cleaning

def clean_text(text: str) -> str:
    text = str(text)
    text = text.lower()
    # remove URLs
    text = re.sub(r"http\S+|www\.\S+", " ", text)
    # remove non-letters (keep basic punctuation)
    text = re.sub(r"[^a-z0-9\s.,!?'-]", " ", text)
    # collapse multiple spaces
    text = re.sub(r"\s+", " ", text).strip()
    return text

# Handle missing values
df["title"] = df["title"].fillna("")
df["description"] = df["description"].fillna("")
df["venue"] = df["venue"].astype(str)

# Combine title + description and clean
df["text_raw"] = df["title"] + " " + df["description"]
df["text"] = df["text_raw"].apply(clean_text)

print("\nSample cleaned text + venue:")
print(df[["text", "venue"]].head())

# 3) Filter rare venues
min_samples = 30
venue_counts = df["venue"].value_counts()
valid_venues = venue_counts[venue_counts >= min_samples].index

df_filtered = df[df["venue"].isin(valid_venues)].copy()
print("\nAfter filtering rare venues:")
print("Filtered shape:", df_filtered.shape)
print("Number of classes (venues):", df_filtered["venue"].nunique())

# 4) Encode labels
label_encoder = LabelEncoder()
df_filtered["label"] = label_encoder.fit_transform(df_filtered["venue"])
num_classes = df_filtered["label"].nunique()

print("Number of classes:", num_classes)
print("\nSome label mappings (venue → id):")
for cls, idx in list(zip(label_encoder.classes_,
                          label_encoder.transform(label_encoder.classes_)))[:10]:
    print(f"{cls} → {idx}")

# 5) Train / Val split
X_text = df_filtered["text"].values
y = df_filtered["label"].values

X_train, X_val, y_train, y_val = train_test_split(
    X_text,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

print("\nTrain size:", len(X_train))
print("Val size:", len(X_val))

# 6) Tokenization & padding
vocab_size = 20000
max_len = 200   # slightly longer than before

tokenizer = Tokenizer(num_words=vocab_size, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train)

X_train_seq = tokenizer.texts_to_sequences(X_train)
X_val_seq = tokenizer.texts_to_sequences(X_val)

X_train_pad = pad_sequences(X_train_seq, maxlen=max_len, padding="post", truncating="post")
X_val_pad = pad_sequences(X_val_seq, maxlen=max_len, padding="post", truncating="post")

print("\nTrain padded shape:", X_train_pad.shape)
print("Val padded shape:", X_val_pad.shape)

# 7) Class weights (for imbalance)
class_weights_array = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(y_train),
    y=y_train
)
class_weights = {i: float(w) for i, w in enumerate(class_weights_array)}
print("\nClass weights:", class_weights)

# 8) Stronger BiLSTM model
#    (with proper Input so summary shows params)
embedding_dim = 128

inputs = layers.Input(shape=(max_len,), dtype="int32", name="input_ids")

x = layers.Embedding(
    input_dim=vocab_size,
    output_dim=embedding_dim,
    name="embedding"
)(inputs)

x = layers.SpatialDropout1D(0.3)(x)

# BiLSTM with sequences
x = layers.Bidirectional(
    layers.LSTM(128, return_sequences=True)
)(x)

# Combine max + average pooling over time
x_max = layers.GlobalMaxPooling1D()(x)
x_avg = layers.GlobalAveragePooling1D()(x)
x = layers.Concatenate()([x_max, x_avg])

x = layers.Dense(128, activation="relu")(x)
x = layers.Dropout(0.5)(x)

outputs = layers.Dense(num_classes, activation="softmax")(x)

model = Model(inputs=inputs, outputs=outputs)

model.compile(
    loss="sparse_categorical_crossentropy",
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    metrics=["accuracy"],
)

print("\nModel summary:")
model.summary()

# 9) Train
batch_size = 64
epochs = 25

es = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    patience=4,
    restore_best_weights=True,
    verbose=1,
)

rlr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=2,
    min_lr=1e-5,
    verbose=1,
)

history = model.fit(
    X_train_pad,
    y_train,
    validation_data=(X_val_pad, y_val),
    batch_size=batch_size,
    epochs=epochs,
    callbacks=[es, rlr],
    class_weight=class_weights,
    verbose=1,
)

# 
# 10) Plot accuracy & loss
plt.figure()
plt.plot(history.history["accuracy"], label="train_acc")
plt.plot(history.history["val_accuracy"], label="val_acc")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.title("BiLSTM: Training vs Validation Accuracy")
plt.show()

plt.figure()
plt.plot(history.history["loss"], label="train_loss")
plt.plot(history.history["val_loss"], label="val_loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.title("BiLSTM: Training vs Validation Loss")
plt.show()

print("Best Validation Accuracy:", max(history.history["val_accuracy"]))

# 11) Prediction helper
def predict_venue(text: str):
    cleaned = clean_text(text)
    seq = tokenizer.texts_to_sequences([cleaned])
    pad = pad_sequences(seq, maxlen=max_len, padding="post", truncating="post")
    probs = model.predict(pad, verbose=0)[0]
    idx = int(np.argmax(probs))
    venue_name = label_encoder.inverse_transform([idx])[0]
    conf = float(probs[idx])
    return venue_name, conf

# Quick sanity check
print("\n=== Some samples from validation set ===")
for i in range(5):
    txt = X_val[i]
    true_label = y_val[i]
    true_venue = label_encoder.inverse_transform([true_label])[0]
    pred_venue, conf = predict_venue(txt)
    print("Text:", txt[:150], "...")
    print("True venue:", true_venue)
    print("Predicted venue:", pred_venue, f"(confidence: {conf:.2f})")
    print("-" * 80)
```
